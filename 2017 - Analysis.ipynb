```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# District 1
df_sector_nilgiri=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Nilgiri\power_consumption_sector_wise_2017_18_4.csv')
df_sector_nilgiri["District"]="The Nilgiris"

# District 2
df_sector_salem=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Salem\power_consumption_sector_wise_2017_18_6.csv')
df_sector_salem['District']="Salem"

# District 3
df_sector_erode=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Erode\power_consumption_sectorwise_2017-18_2.csv')
df_sector_erode['District']="Erode"

# District 4
df_sector_kancheepuram=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Kancheepuram\power_consumption_sectorwise_2017_18_4.csv')
df_sector_kancheepuram['District']='Kancheepuram'

# District 5
df_sector_tirunelveli=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Thirunelveli\power_consumption_sector_wise_2017_18_5.csv')
df_sector_tirunelveli['District']='Tirunelveli'

# District 6
df_sector_cuddalore=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Cuddalore\power_consumption_sectorwise_2017_18_5.csv')
df_sector_cuddalore['District']='Cuddalore'

# District 7
df_sector_sivagangai=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Sivagangai\sectorwise_power_consumption_2017_18.csv')
df_sector_sivagangai['District']='Sivaganga'

# District 8
df_sector_nagapatinam=pd.read_csv(r'C:\Users\Praveen Gandhi\Documents\Athena Infonomics\2017\Nagapatinam\power_consumption_sector_wise_2017-18_2.csv')
df_sector_nagapatinam['District']='Nagappattinam'
```

## District 1 The Nilgiri

```python
df_sector_nilgiri
```

```python
df_sector_nilgiri=df_sector_nilgiri.rename(columns={'Consumption (in Million Units)':'Consumption(MU)','Percentage of the Consumption':'Percentage'})
df_sector_nilgiri=df_sector_nilgiri[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_nilgiri= df_sector_nilgiri.replace(np.nan, 0)
df_sector_nilgiri['Consumption(MU)']=df_sector_nilgiri['Consumption(MU)'].astype(float)
df_sector_nilgiri['Percentage']=df_sector_nilgiri['Percentage'].astype(int)

df_sector_nilgiri1=df_sector_nilgiri.copy()
df_sector_nilgiri= df_sector_nilgiri[:-1]

# plot 
display(plt.plot(df_sector_nilgiri['Consumption(MU)'], df_sector_nilgiri['Sectors']))
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("The Nilgiris District Sector wise power consumption 2017-18")

print(df_sector_nilgiri1[df_sector_nilgiri1['Sectors']=='Total'])
```

```python
# plot 
plt.figure(figsize=(16,6))
display(plt.plot(df_sector_nilgiri['Sectors'],df_sector_nilgiri['Percentage']))
plt.ylabel('Power Consumptions(MU) %') 
plt.xlabel('Sectors') 
  
# displaying the title
plt.title("Nilgiri District Sector wise power consumption % 2017-18")
```

## District 3 Erode

```python
df_sector_erode
```

```python
df_sector_erode=df_sector_erode.rename(columns={'Total District Consumption (M.U)':'Consumption(MU)'})
df_sector_erode= df_sector_erode.replace(np.nan, 0)
df_sector_erode= df_sector_erode.replace('NIL', 0)


df_sector_erode['Percentage of Electricity Consumption in Erode Division']=df_sector_erode['Percentage of Electricity Consumption in Erode Division'].astype(float)
df_sector_erode['Percentage of Electricity Consumption in Gobichettipalayam Division']=df_sector_erode['Percentage of Electricity Consumption in Gobichettipalayam Division'].astype(float)
df_sector_erode['Percentage']=(df_sector_erode['Percentage of Electricity Consumption in Erode Division']+df_sector_erode['Percentage of Electricity Consumption in Gobichettipalayam Division'])/2
df_sector_erode=df_sector_erode[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_erode['Consumption(MU)']=df_sector_erode['Consumption(MU)'].astype(float)
df_sector_erode['Percentage']=df_sector_erode['Percentage'].astype(int)

df_sector_erode1=df_sector_erode.copy()
df_sector_erode= df_sector_erode[:-1]

# Prints Total consumption in MU
print(df_sector_erode1[df_sector_erode1['Sectors']=='Total'])

# Plotting
plt.plot(df_sector_erode['Consumption(MU)'], df_sector_erode['Sectors'])
  
# changing the size of figure to 2X2
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("Erode District Sector wise power consumption 2017-18")
```

## District 4 Kancheepuram

```python
df_sector_kancheepuram
```

```python
df_sector_kancheepuram= df_sector_kancheepuram.replace(np.nan, 0)
df_sector_kancheepuram= df_sector_kancheepuram.replace('Unmetered', 0)

df_sector_kancheepuram['Percentage of the Consumption']=df_sector_kancheepuram['Percentage of the Consumption'].astype(float)
df_sector_kancheepuram['Percentage of the Consumption.1']=df_sector_kancheepuram['Percentage of the Consumption.1'].astype(float)
df_sector_kancheepuram['Percentage of the Consumption.2']=df_sector_kancheepuram['Percentage of the Consumption.2'].astype(float)

df_sector_kancheepuram['Kancheepuram Division Consumption (in MU)']=df_sector_kancheepuram['Kancheepuram Division Consumption (in MU)'].astype(float)
df_sector_kancheepuram['Chengalpattu Division Consumption (in MU)']=df_sector_kancheepuram['Chengalpattu Division Consumption (in MU)'].astype(float)
df_sector_kancheepuram['KK Nagar Division Consumption (in MU)']=df_sector_kancheepuram['KK Nagar Division Consumption (in MU)'].astype(float)

df_sector_kancheepuram['Percentage of the Consumption']=df_sector_kancheepuram['Percentage of the Consumption'].astype(int)
df_sector_kancheepuram['Percentage of the Consumption.1']=df_sector_kancheepuram['Percentage of the Consumption.1'].astype(int)
df_sector_kancheepuram['Percentage of the Consumption.2']=df_sector_kancheepuram['Percentage of the Consumption.2'].astype(int)

df_sector_kancheepuram['Consumption(MU)']=df_sector_kancheepuram['Kancheepuram Division Consumption (in MU)']+df_sector_kancheepuram['Chengalpattu Division Consumption (in MU)']+df_sector_kancheepuram['KK Nagar Division Consumption (in MU)']
df_sector_kancheepuram['Percentage']=(df_sector_kancheepuram['Percentage of the Consumption']+df_sector_kancheepuram['Percentage of the Consumption.1']+df_sector_kancheepuram['Percentage of the Consumption.2'])/3
df_sector_kancheepuram=df_sector_kancheepuram[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_kancheepuram1=df_sector_kancheepuram.copy()
df_sector_kancheepuram= df_sector_kancheepuram[:-1]

# Prints Total consumption in MU
print(df_sector_kancheepuram1[df_sector_kancheepuram1['Sectors']=='Grand Total'])

# Plotting
plt.plot(df_sector_kancheepuram['Consumption(MU)'], df_sector_kancheepuram['Sectors'])
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("Kancheepuram District Sector wise power consumption 2017-18")
```

```python
# plot 
plt.figure(figsize=(16,6))
plt.plot(df_sector_kancheepuram['Sectors'],df_sector_kancheepuram['Percentage'])
plt.ylabel('Power Consumptions(MU) %') 
plt.xlabel('Sectors') 
  
# displaying the title
plt.title("Kancheepuram District Sector wise power consumption % 2017-18")
plt.show()
```
