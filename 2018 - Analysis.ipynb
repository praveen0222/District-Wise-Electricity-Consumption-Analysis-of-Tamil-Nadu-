```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# District 1
df_sector_nilgiri=pd.read_csv('nilgiri.csv')
df_sector_nilgiri["District"]="The Nilgiris"

# District 2
df_sector_salem=pd.read_csv('salem.csv')
df_sector_salem['District']="Salem"

# District 3
df_sector_erode=pd.read_csv('erode.csv')
df_sector_erode['District']="Erode"

# District 4
df_sector_tirunelveli=pd.read_csv('thirunelveli.csv')
df_sector_tirunelveli['District']='Tirunelveli'

# District 5
df_sector_cuddalore=pd.read_csv('cuddalore.csv')
df_sector_cuddalore['District']='Cuddalore'
```

## District 1 The Nilgiris

```python
df_sector_nilgiri
```

```python
df_sector_nilgiri=df_sector_nilgiri.rename(columns={'Power Consumption (in MU)':'Consumption(MU)','Percentage of the Consumption':'Percentage'})
df_sector_nilgiri=df_sector_nilgiri[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_nilgiri= df_sector_nilgiri.replace(np.nan, 0)
df_sector_nilgiri['Consumption(MU)']=df_sector_nilgiri['Consumption(MU)'].astype(float)
df_sector_nilgiri['Percentage']=df_sector_nilgiri['Percentage'].astype(int)

df_sector_nilgiri1=df_sector_nilgiri.copy()
df_sector_nilgiri= df_sector_nilgiri[:-1]

# plot 
display(plt.plot(df_sector_nilgiri['Consumption(MU)'], df_sector_nilgiri['Sectors']))
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("The Nilgiris District Sector wise power consumption 2018-2019")

print(df_sector_nilgiri1[df_sector_nilgiri1['Sectors']=='Total'])
```

```python
# plot 
plt.figure(figsize=(16,6))
display(plt.plot(df_sector_nilgiri['Sectors'],df_sector_nilgiri['Percentage']))
plt.ylabel('Power Consumptions(MU) %') 
plt.xlabel('Sectors') 
  
# displaying the title
plt.title("Nilgiri District Sector wise power consumption % 2018-19")
```

## District 2 Salem

```python
df_sector_salem
```

```python
df_sector_salem=df_sector_salem.rename(columns={'Total Power Consumption (in Million Units)':'Consumption(MU)',
                                                'Total Percentage of Power Consumption':'Percentage'})
df_sector_salem['Percentage']=df_sector_salem['Percentage']/2
df_sector_salem=df_sector_salem[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_salem= df_sector_salem.replace(np.nan, 0)
df_sector_salem['Consumption(MU)']=df_sector_salem['Consumption(MU)'].astype(float)
df_sector_salem['Percentage']=df_sector_salem['Percentage'].astype(int)

df_sector_salem1=df_sector_salem.copy()
df_sector_salem= df_sector_salem[:-1]

# Plot total consumption
print(df_sector_salem1[df_sector_salem1['Sectors']=='Total'])

# Plotting
display(plt.plot(df_sector_salem['Consumption(MU)'], df_sector_salem['Sectors']))
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("Salem District Sector wise power consumption 2018-19")
```

```python
# plot 
plt.figure(figsize=(16,6))
plt.plot(df_sector_salem['Sectors'],df_sector_salem['Percentage'])
plt.ylabel('Power Consumptions(MU) %') 
plt.xlabel('Sectors') 
  
# displaying the title
plt.title("Salem District Sector wise power consumption % 2018-19")
plt.show()
```

## District 3 Erode

```python
df_sector_erode
```

```python
df_sector_erode=df_sector_erode.rename(columns={'Total Consumption in District (MU)':'Consumption(MU)'})
df_sector_erode= df_sector_erode.replace(np.nan, 0)
df_sector_erode= df_sector_erode.replace('NIL', 0)


df_sector_erode['Percentage of the Consumption in Erode Electricity Division']=df_sector_erode['Percentage of the Consumption in Erode Electricity Division'].astype(float)
df_sector_erode['Percentage of the Consumption in Gobichettipalayam Electricity Division']=df_sector_erode['Percentage of the Consumption in Gobichettipalayam Electricity Division'].astype(float)
df_sector_erode['Percentage']=(df_sector_erode['Percentage of the Consumption in Erode Electricity Division']+df_sector_erode['Percentage of the Consumption in Gobichettipalayam Electricity Division'])/2
df_sector_erode=df_sector_erode[['Sectors','Consumption(MU)','Percentage','District']]

df_sector_erode['Consumption(MU)']=df_sector_erode['Consumption(MU)'].astype(float)
df_sector_erode['Percentage']=df_sector_erode['Percentage'].astype(int)

df_sector_erode1=df_sector_erode.copy()
df_sector_erode= df_sector_erode[:-1]

# Prints Total consumption in MU
print(df_sector_erode1[df_sector_erode1['Sectors']=='Total'])

# Plotting
plt.plot(df_sector_erode['Consumption(MU)'], df_sector_erode['Sectors'])
  
# changing the size of figure to 2X2
plt.xlabel('Power Consumptions(MU)') 
plt.ylabel('Sectors') 
  
# displaying the title
plt.title("Erode District Sector wise power consumption 2018-19")
```

```python
# plot 
plt.figure(figsize=(18,6))
plt.plot(df_sector_erode['Sectors'],df_sector_erode['Percentage'])
plt.ylabel('Power Consumptions(MU) %') 
plt.xlabel('Sectors') 
  
# displaying the title
plt.title("Erode District Sector wise power consumption % 2018-19")
plt.show()
```
